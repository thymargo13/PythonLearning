## Merge
pd.merge(left, right,
        how = 'inner',
        on=None,
        left_on=None,
        right_on =None,
        left_index = False,
        right_index = False,
        sort = True,
        suffixes('_x','_y'),
        copy =True,
        indicator = False,
        validate = None)
        
#### can only merge two table, one is right , another is left
#### how means the merge method.
#### on means to follow a specific column to merge

```python
left = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'A':['A0','A1','A2','A3'],
                    'B':['B0','B1','B2','B3']})
left
```

```python
right = pd.DataFrame({'key':['K0','K1','K2','K3'],
                    'C':['C0','C1','C2','C3'],
                    'D':['D0','D1','D2','D3']})
right
```

```python
pd.merge(left,right,on='key') # on is their common column
```

```python
left = pd.DataFrame({'firstname':['Pool','House','Mine','Will'],
                    'location':[1,2,1,4]})
left
```

```python
right = pd.DataFrame({'lastname':['Henry','Jones','Indian','Jack'],
                    'location':[1,2,3,5]})
right
```

```python
pd.merge(left,right,on ='location')
```

```python
pd.merge(left,right,on ='location',how='outer') # default is inner
```

```python
pd.merge(left,right,on ='location', how ='left')
```

```python
pd.merge(left,right,on ='location', how ='right')
```

```python
left = pd.DataFrame({'firstname':['Pool','House','Mine','Will']},
                    index = ['a','a','b','b'])
left
```

```python
right = pd.DataFrame({'lastname':['Henry','Jones','Indian','Jack'],
                    'location':['a','b','c','d']})
right
```

```python
pd.merge(left,right,left_index = True, right_on='location')
# left_index --> using left index 
# and right location column
```

```python
#join(self, other, on=None, how='left', lsuffix='', rsuffix='',sort=False):
#通过lsuffix='', rsuffix='' 区分相同列名的列
d3_2 = d1.join(d2.set_index(1, drop = False),how='outer', on=['E'],) # drop = False for keep this column
d3_2
```

```python
d3 = d3_2.dropna(how='any')
d3.index = (['01','02','03','04'])
d3
```

## Exercise 3

```python
L=[]
for i in d3.index:
    if d3.loc[i,2] == 'S':
        p = d3.loc[i,'B'] + d3.loc[i,'C'] - d3.loc[i,3]
        L.append(p)
    if d3.loc[i,2] == 'T':
        p = d3.loc[i,'B'] - d3.loc[i,'C'] + d3.loc[i,3]
        L.append(p)
d4 = pd.DataFrame(L)
result = pd.concat([d3[2].reset_index(drop=True),d4], axis =1, ignore_index = True)
result.columns = ['Type','Value']
result
```
