## ax.annotate

```python
fig = plt.figure(figsize= (10,7))
ax = fig.add_subplot(1,1,1)
x = df2.close[:80]
y = df1.close[:80]
plt.scatter(x,y)
plt.title('GOOG vs AAPL')
plt.xlabel('GOOG')
plt.ylabel('APPL')

ax.annotate(str(df2.index[10])[:10],(x[10],y[10])) #(context,location)

plt.show()
```

```python
fig = plt.figure(figsize= (10,7))
ax = fig.add_subplot(1,1,1)
x = df2.close[:80]
y = df1.close[:80]
plt.scatter(x,y)
plt.title('GOOG vs AAPL')
plt.xlabel('GOOG')
plt.ylabel('APPL')

for i in range(80):
    ax.annotate(str(df2.index[i])[:10],(x[i],y[i])) #(context,location)

plt.show()
```

```python
# AAPL stock info from 22-10-2018 to 26-10-2018
df1 = pd.read_csv('AAPL_intraday_data.csv')
df1.head()
df1.columns = ['Time', 'Open', 'High', 'Low', 'Close', 'Volume']
```

```python
print(len(df1))
day1 = df1[:390]
print(len(day1))
day1.head()
```

## Excerise 2

### Method 1
using plt.xticks and set range intervals.

```python
k = list(range(0,390,35))
k
```

```python
plt.figure(figsize=(10,5))
plt.plot(day1.index,day1.Close) 
plt.title("Apple Ltd. close price in 22-10-2018")
plt.xticks(day1.index[k], day1.Time[k])
plt.show()
```

### Method 2
##### datetime method.
%Y -- year, %d -- day, %m -- Month, %H -- Hour, %M -- minutes, %S -- second


```python
type(day1.Time[0])
```

```python
day1.Time = pd.to_datetime(day1.Time,format='%H:%M:%S')
day1.head()
```

```python
# pd.to_datetime(data, format ='').dt.time --> in time format
day1.Time = pd.to_datetime(day1.Time,format='%H:%M:%S').dt.time
day1.head()
```

```python
# pd.to_datetime(data, format ='').dt.date --> in date format
day1.Time = pd.to_datetime(day1.Time,format='%H:%M:%S').dt.date
day1.head()
```

```python
plt.figure(figsize=(10,5))
plt.plot(day1.Time,day1.Close) 
plt.title("Apple Ltd. close price in 22-10-2018")
plt.show()
```

```python
day1.head()
```

```python
def difference(high, low):
    difference= high - low
    return difference

d = difference(day1.High, day1.Low) #type : Series

day1.index = day1.Time
fig = plt.figure(figsize=(20,10))
 #(number of chart, no. of each row, position)
ax = fig.add_subplot(2,2,1)
day1.Open.plot(label ='High')
plt.legend()
plt.title('Intraday Price')
plt.ylabel('Price')

ax = fig.add_subplot(2,2,1)
day1.Close.plot(label ='Low')
plt.legend()
plt.ylabel('Price')
plt.title('Intraday Price')

ax = fig.add_subplot(2,2,2)
plt.bar(d.index, d.values, width=30)# (x,y)
plt.ylabel('Price')
plt.title('Intraday Price Difference')

plt.show()
```
